```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.constants as sc
import sympy as smp
import matplotlib.animation as animation


u,v,vectorfield,X,Y,V1,V2=smp.symbols("u v vectorfield X Y V1 V2",cls=smp.Function)
r=smp.symbols("r",cls=smp.Function)
t=smp.symbols("t")

X=X(t)# curve paramaters
Y=Y(t)
V1=V1(X,Y) #vect components where x and y of curve is inputted
V2=V2(X,Y)
vectorfield=smp.Matrix([V1,V2])#vectorfield where pareameter is set already
r= smp.Matrix([X,Y])#parameteriazation
```

Line integral can be written as $\int F(x,y).dl= \int_{b}^{a}F(\vec{r}(t)).\vec{r(t)}'dt$.where $\vec{r}(t)$ is a parametric function along a path $b\leq t\leq a$, and $\vec{r}(t) =\begin{bmatrix}
x(t)\\
y(t) \end{bmatrix}$ and, $F(x,y)$ is a 2d vector field such that $\vec{F}(x,y)=\begin{bmatrix}
V1(x,y)\\
V2(x,y) \end{bmatrix}$

```python
integrand = vectorfield.dot(smp.diff(r,t))

integrand
```

```python
U,V=smp.cos(t)+2, smp.sin(t) #parametric functions
smp.Matrix([U,V])
```

```python
vectorfield1,vectorfield2 =-Y,X  #plugging in parametric to vector field
smp.Matrix([vectorfield1,vectorfield2])
```

```python
integrationfact= integrand.subs([(V1,vectorfield1),(V2,vectorfield2),(X,U),(Y,V)]).doit().simplify()
integrationfact
```

```python
smp.integrate(integrationfact,(t,0,2*smp.pi))
```

```python

plt.title('Graph for line intergral')
z= np.linspace(0,2*sc.pi, 1200)


u,v=np.cos(z)+2, np.sin(z)#parametric for graph

x, y = np.meshgrid(np.arange(-5,5,0.2),
                   np.arange(-5,5,0.2))
x1=-y
y1=x#for field
mag=np.sqrt(x1**2+y1**2)

plt.quiver(x, y,x1,y1,mag,cmap="jet",scale_units="inches",scale=10,animated=True)


#parametric fun


plt.plot(u,v,label='parametric fun')
plt.legend()



plt.grid()
ax = plt.gca()
ax.set_xlim([0,4])
ax.set_ylim([-2,2])

plt.show()




```
